```python
import pandas as pd
movie_data = pd.read_csv('/movie_data.csv')
rating_data = pd.read_csv('/rating.txt')
```

```python
rating_data.head()
```

```python
movie_data.head()
```

```python
print(movie_data.shape)
print(rating_data.shape)
```

```python
rating_data.drop('id', axis=1, inplace=True)
rating_data.head()
```

```python
movie_data.rename(columns = {'5' : 'movie_id'}, inplace = True)
movie_data.rename(columns = {'포룸' : 'title'}, inplace = True)
movie_data.rename(columns = {'1995-12-09' : 'release_date'}, inplace = True)
data_to_insert = {'movie_id': 5, 'title': '포룸', 'release_date': '1995-12-09'}
movie_data = movie_data.append(data_to_insert, ignore_index = True)
movie_data.head()
```

```python
user_movie_data = pd.merge(rating_data, movie_data, on='movie_id')
user_movie_data.head()
```

```python
user_movie_data.shape
```

```python
user_movie_rank=user_movie_data.pivot_table('rank',index='user_id', columns='title').fillna(0)
```

```python
user_movie_rank.shape
```

```python
user_movie_rank
```

```python
movie_user_rank = user_movie_rank.values.T
```

```python
movie_user_rank
```

```python
from sklearn.decomposition import TruncatedSVD
SVD = TruncatedSVD(n_components=12)
matrix = SVD.fit_transform(movie_user_rank)
```

```python
matrix[0]
```

```python
import numpy as np
corr = np.corrcoef(matrix)
```

```python
corr.shape
```

```python
movie_title = user_movie_rank.columns
movie_title_list = list(movie_title)
coffey_hands = movie_title_list.index("러브 라이브! 더 스쿨 아이돌 무비")
```

```python
corr_coffey_hands = corr[coffey_hands]
list(movie_title[(corr_coffey_hands) >= 0.9])[:50]
```
